```python
import folium
```

```python
latitude=12.9726
longitude=77.5946
m=folium.Map(location=[latitude,longitude],zoom_start=12)
```

```python
locations = (
    {"name": "bangalore palace", "latitude": 12.9958, "longitude": 77.58},
    {"name": "vidhana soudha", "latitude": 12.9722, "longitude": 77.59},
    {"name": "indian institute of science", "latitude": 12.9984, "longitude": 77.56},
)
```

```python
for location in locations:
  folium.Marker(
      [location["latitude"], location["longitude"]], popup=location["name"]
  ).add_to(m)
m
```
