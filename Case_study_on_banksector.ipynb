```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


age_groups = ['10-21', '22-35', '36-50', '51-60', '61-65', '66+']


total_loans = np.random.randint(50, 200, size=len(age_groups))
bad_loans = np.random.randint(0, total_loans)
good_loans = total_loans - bad_loans
percent_bad_loans = (bad_loans / total_loans) * 100
percent_good_loans = 100 - percent_bad_loans


data = {
    'Age Group': age_groups,
    'Total Loans': total_loans,
    'Bad Loans': bad_loans,
    'Good Loans': good_loans,
    'Percentage Bad Loans': percent_bad_loans,
    'Percentage Good Loans': percent_good_loans
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.bar(df['Age Group'], df['Total Loans'], color='pink')
plt.title('Total Loans')
plt.xlabel('Age Group')
plt.ylabel('Number of Loans')
```

```python

labels = data['Age Group']
sizes = data['Percentage Good Loans']
colors = ['pink', '#99ff99']
explode = (0, 0, 0, 0, 0, 0.1)

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Loans by Age Group')
plt.show()
```

```python

plt.hist([good_loans, bad_loans], bins=range(0, 220, 20), alpha=0.5, label=['Good Loans', 'Bad Loans'])
plt.xlabel('Number of Loans')
plt.ylabel('Frequency')
plt.title('Histogram of Good and Bad Loans')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

age_groups = ['10-21', '22-35', '36-50', '51-60', '61-65', '66+']
total_loans = np.random.randint(50, 200, size=len(age_groups))
bad_loans = np.random.randint(0, total_loans)
good_loans = total_loans - bad_loans


good_loans_norm = good_loans / total_loans
bad_loans_norm = bad_loans / total_loans
fig, ax = plt.subplots()
ax.bar(age_groups, good_loans_norm, label='Good Loans')
ax.bar(age_groups, bad_loans_norm, bottom=good_loans_norm, label='Bad Loans')


ax.set_title('Normalized Good and Bad Loans by Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Normalized Loans')
ax.legend()
plt.show()
```
